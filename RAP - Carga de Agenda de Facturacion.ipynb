```python
def get_ultimo_dia_del_mes(fecha):
    siguiente_mes = fecha.replace(day=28) + datetime.timedelta(days=4)  # this will never fail
    return siguiente_mes - datetime.timedelta(days=siguiente_mes.day)
```

```python
def get_numero_dia_de_semana(fecha):
    #numero_dia_semana = fecha.weekday()
    
    #if (numero_dia_semana == 7):
    #    numero_dia_semana = 0
    #else:
    #    numero_dia_semana = numero_dia_semana + 1
    
    numero_dia_semana = datetime.datetime.isocalendar(fecha)[2]
    
    if (numero_dia_semana == 7):
        numero_dia_semana = 0
    
    return numero_dia_semana
```

```python
def maximizar_pantalla():
    r.keyboard('[alt][space]')
    r.keyboard('x')
    r.wait(1)
```

```python
def seleccionar_fecha(fecha):
    maximizar_pantalla()
    
    fila, columna = get_fila_columna_calendario(fecha)
    
    clicks = get_click_mes_siguiente(date.today(), fecha)
    
    if (clicks > 0):
        for i in range(clicks):
            # Siguiente Mes
            r.click(calendario_boton_siguiente_x, calendario_boton_siguiente_y)
            r.wait(2)
            
    r.click(calendario_fecha_Proceso_columnas[columna], calendario_fecha_Proceso_filas[fila])
    
    r.wait(1)
```

```python
#2022-01-03
def get_fila_columna_calendario(fecha):
    #numero_semana = get_numero_semana(fecha)
    #numero_semana_primera = get_numero_semana(datetime.datetime(fecha.year,fecha.month,1))
    
    #fila = numero_semana - numero_semana_primera
    
    fila = get_numero_semana(fecha)
    
    columna = get_numero_dia_de_semana(fecha)
    
    return fila, columna
```

```python
def get_click_mes_siguiente(fecha_pivot, fecha):
    
    diferencia = fecha.month - fecha_pivot.month
    
    if(fecha.year > fecha_pivot.year):
        diferencia = (12 + fecha.month) - fecha_pivot.month
    
    return diferencia
```
